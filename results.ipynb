# ESGF Virtual Aggregation

The aim of this project is to create a ready-to-deploy TDS catalog including ALL available data in ESGF, using OPeNDAP endpoints to provide ESGF data analysis while avoiding the download of any data from remote repositories.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
df = pd.read_csv("multiple-runs-48-compression10.csv")
```

### Analyze the results

```python
g = sns.FacetGrid(df[df['name'] == 'Kerchunk'], col="data_node", col_wrap=2, height=3.5, aspect=1.5)
g.map_dataframe(sns.barplot, x="name", y="time", hue="workers")
g.add_legend(title="Workers")

plt.savefig("kerchunk-time.svg")
plt.show()
```

```python
g = sns.FacetGrid(df[df['name'] == 'Kerchunk'], col="data_node", col_wrap=2, height=3.5, aspect=1.5)
g.map_dataframe(sns.barplot, x="name", y="bytes_recv", hue="workers")
g.add_legend(title="Workers")

plt.savefig("kerchunk-bytes.svg")
plt.show()
```
